```python
import pandas as pd

from dash.dependencies import Input, Output
import dash_core_components as dcc
import dash_html_components as html
import plotly.graph_objs as go
import dash_table_experiments as dt

from sklearn.cluster import KMeans

import sqlite3
import re
```

```python
data = pd.read_csv('../../data/example-1.csv', index_col = 0)
```

```python
data.head()
```

```python
X = data
```

```python
kmeans = KMeans(n_clusters=15, max_iter=300, random_state=4)
kmeans.fit(X)

labels_kmeans = kmeans.labels_
df_clusterid = pd.DataFrame(labels_kmeans, index=X.index)
df_clusterid.rename(columns={0: "cluster"}, inplace=True)
df_clusters = pd.concat([X, df_clusterid], axis=1)
```
